```python
import matplotlib.pyplot as plt                  # For plotting
import netCDF4                                   # For reading data
pylab.rcParams['figure.figsize'] = (17.0, 10.0)  # Large figures
path = './WSwPSBF.A1m/'
```

```python
visc = netCDF4.Dataset(path+'visc.nc')
for v in visc.variables: print v,

# recall data layout is (t,z,y,x)

time = visc.variables['Time'][:]

# KPP boundary layer depth as function of time (metre)
h    = visc.variables['KPP_OBLdepth'][:,0,0]

# KPP tracer diffusivity as function of time and depth (m2/sec)
Kt   = visc.variables['KPP_Kheat'][:,:,0,0]

# KPP velocity diffusivity as function of time and depth (m2/sec)
Ku   = visc.variables['KPP_Kv'][:,:,0,0]

# surface (and penetrating) buoyancy flux, as used by [CVmix] KPP (m2/s3)
KPP_buoyFlux = visc.variables['KPP_buoyFlux'][:,:,0,0]

# Temperature tendency due to non-local transport of heat, as calculated by KPP (K/s)
KPP_dTdt = visc.variables['KPP_dTdt'][:,:,0,0]
```

```python
prog  = netCDF4.Dataset(path+'prog.nc')
for v in prog.variables: print v,
   
# depth of cell interface     
zi = prog.variables['zi'][:]

# depth of cell center 
zl = prog.variables['zl'][:]

# zonal velocity as function of time and depth
u  = prog.variables['u'][:,:,0,0]

# zonal velocity as function of time and depth
v  = prog.variables['v'][:,:,0,0]

# temperature as function of time and depth 
temp = prog.variables['temp'][:,:,0,0]
```

```python
# KPP related fields 

plt.subplot(2,2,1)
plt.contourf(time, -zi, Kt.T, levels=numpy.arange(0,.04,.005))
plt.colorbar()
plt.ylim((-70,0))
plt.ylabel('z (m)',fontsize=14)
plt.title(r'$\kappa_\t (m^2/s)$',fontsize=18)

plt.subplot(2,2,2)
plt.contourf(time, -zi, Ku.T, levels=numpy.arange(0,.04,.005))
plt.colorbar()
plt.ylim((-70,0))
plt.ylabel('z (m)',fontsize=14)
plt.title(r'$\kappa_u (m^2/s)$',fontsize=18)
plt.xlabel('Time (days)',fontsize=14)

plt.subplot(2,2,3)
plt.contourf(time, -zi, KPP_buoyFlux.T, 16)
plt.colorbar()
plt.ylim((-70,0))
plt.ylabel('z (m)',fontsize=14)
plt.title(r'$Buoyancy flux (m^2/s^3)$',fontsize=18)
plt.xlabel('Time (days)',fontsize=14)

plt.subplot(2,2,4)
plt.contourf(time, -zl, KPP_dTdt.T, 16)
plt.colorbar()
plt.ylim((-70,0))
plt.ylabel('z (m)',fontsize=14)
plt.title('Temp tendency from non-local (K/s)',fontsize=18)
plt.xlabel('Time (days)',fontsize=14);
```

```python
# Boundary layer depth and SST 

plt.subplot(2,1,1)
plt.plot(time, -h,'-')
plt.ylim((-100,0))
plt.grid()
plt.ylabel('z (m)',fontsize=14)
plt.title('KPP boundary layer depth for WSwPSBF.A',fontsize=16)

plt.subplot(2,1,2)
plt.plot(time, temp[:,0],'-')
plt.ylim((14,15.0))
plt.grid()
plt.xlabel('Time (days)',fontsize=14)
plt.ylabel(r'$\theta$ (${\degree}C$)',fontsize=14)
plt.title('SST for WSwPSBF.A',fontsize=16);
```

```python
# u,v,temp

plt.subplot(2,2,1)
plt.contourf(time, -zl, u.T, levels=numpy.arange(-.10,.11,.01))
plt.colorbar()
plt.ylim((-60,0))
plt.ylabel('z (m)',fontsize=14)
plt.title('u (m/s)',fontsize=16)

plt.subplot(2,2,2)
plt.contourf(time, -zl, v.T, levels=numpy.arange(-.10,.11,.01))
plt.colorbar()
plt.ylim((-60,0))
plt.ylabel('z (m)',fontsize=14)
plt.title('v (m/s)',fontsize=16)

plt.subplot(2,2,3)
plt.contourf(time, -zl, temp.T, levels=numpy.arange(14.5,15,.01) )
plt.colorbar()
plt.ylim((-60,0))
plt.ylabel('z (m)',fontsize=14)
plt.title(r'$\theta$ ($\degree$C)',fontsize=16)
plt.xlabel('Time (days)',fontsize=14)


plt.subplot(2,2,4)
difftemp = temp[:,:] - temp[0,:]
CS = plt.contourf(time, -zl, difftemp.T,levels=numpy.arange(-.25,.25,.02))
plt.colorbar()
#C = plt.contour(time, -zl, difftemp.T,8,linewidth=.5, colors='black')
C = plt.contour(time, -zl, difftemp.T, 8,linewidth=.5, colors='black')
plt.clabel(C, inline=1, fontsize=10)
plt.ylim((-60,0))
plt.ylabel('z (m)',fontsize=14)
plt.title(r'$\theta-\theta(t=0)$ ($\degree$C)',fontsize=16);
plt.xlabel('Time (days)',fontsize=14);

```
