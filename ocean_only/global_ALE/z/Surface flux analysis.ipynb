```python
%pylab inline
import matplotlib.pyplot as plt
import netCDF4
```

```python
path    = './'
static  = netCDF4.Dataset(path+'ocean_static.nc')
forcing = netCDF4.Dataset(path+'forcing.nc')
surface = netCDF4.Dataset(path+'surface.nc') ; tvar = 'Time'
cp   = 3925.0
rho0 = 1035.0
n    = 4

lon  = static.variables['geolon'][:]
lat  = static.variables['geolat'][:]
wet  = static.variables['wet'][:]
area = static.variables['area_t'][:]*wet

for v in forcing.variables: print v,
for v in surface.variables: print v,
for v in static.variables: print v,    
```

```python
# for easy setup of subplots
def newSP(y,x):
    global __spv, __spi ; __spv = (y,x) ; __spi = 1 ; plt.subplot(__spv[0], __spv[1], __spi)
def nextSP():
    global __spv, __spi ; __spi = __spi + 1 ; plt.subplot(__spv[0], __spv[1], __spi)
```

<h2 align="center">Basic components of surface heat flux</h2> 

```python
# Basic components of surface heat flux 
# net_heat_surface = heat_pme + frazil + net_heat_coupler + heat_restore

net_heat_surface = forcing.variables['net_heat_surface'][n]
net_heat_coupler = forcing.variables['net_heat_coupler'][n]
heat_pme         = forcing.variables['Heat_PmE'][n]
frazil           = forcing.variables['frazil'][n]
heat_restore     = forcing.variables['heat_restore'][n]

figure(figsize=(18,12))
newSP(3,2);

field     = net_heat_surface
field_min = numpy.amin(field)
field_max = numpy.amax(field)
field_ave = (field*area).sum() / area.sum()
ch = plt.pcolormesh(lon,lat,field)
cbax=plt.colorbar(ch, extend='both')
plt.title('net_heat_surface[$W/m^2$]') 
plt.clim(-200,200)
plt.xlim(-280,80)
plt.ylim(-80,90)
plt.ylabel(r'Latitude [$\degree$N]')
#plt.xlabel(r'Longitude')
plt.gca().set_axis_bgcolor('gray')
axis = plt.gca()
axis.annotate('max=%5.2f\nmin=%5.2f\nave=%5.2f'%(field_max,field_min,field_ave),xy=(0.01,0.73), 
              xycoords='axes fraction', verticalalignment='bottom', fontsize=8, color='black')


nextSP() 
field     = net_heat_coupler
field_min = numpy.amin(field)
field_max = numpy.amax(field)
field_ave = (field*area).sum() / area.sum()
ch = plt.pcolormesh(lon,lat,field)
cbax=plt.colorbar(ch, extend='both')
plt.title('net_heat_coupler[$W/m^2$]') 
plt.clim(-200,200)
plt.xlim(-280,80)
plt.ylim(-80,90)
plt.ylabel(r'Latitude [$\degree$N]')
#plt.xlabel(r'Longitude')
plt.gca().set_axis_bgcolor('gray')
axis = plt.gca()
axis = plt.gca()
axis.annotate('max=%5.2f\nmin=%5.2f\nave=%5.2f'%(field_max,field_min,field_ave),xy=(0.01,0.73), 
              xycoords='axes fraction', verticalalignment='bottom', fontsize=8)

nextSP() 
field     = frazil
field_min = numpy.amin(field)
field_max = numpy.amax(field)
field_ave = (field*area).sum() / area.sum()
ch = plt.pcolormesh(lon,lat,field)
cbax=plt.colorbar(ch, extend='both')
plt.title('frazil[$W/m^2$]') 
plt.clim(-200,200)
plt.xlim(-280,80)
plt.ylim(-80,90)
plt.ylabel(r'Latitude [$\degree$N]')
#plt.xlabel(r'Longitude')
plt.gca().set_axis_bgcolor('gray')
axis = plt.gca()
axis.annotate('max=%5.2f\nmin=%5.2f\nave=%5.2f'%(field_max,field_min,field_ave),xy=(0.01,0.73), 
              xycoords='axes fraction', verticalalignment='bottom', fontsize=8)

nextSP() 
field     = heat_pme
field_min = numpy.amin(field)
field_max = numpy.amax(field)
field_ave = (field*area).sum() / area.sum()
ch = plt.pcolormesh(lon,lat,field)
cbax=plt.colorbar(ch, extend='both')
plt.title('heat_pme[$W/m^2$]') 
plt.clim(-20,20)
plt.xlim(-280,80)
plt.ylim(-80,90)
plt.ylabel(r'Latitude [$\degree$N]')
#plt.xlabel(r'Longitude')
plt.gca().set_axis_bgcolor('gray')
axis = plt.gca()
axis.annotate('max=%5.2f\nmin=%5.2f\nave=%5.2f'%(field_max,field_min,field_ave),xy=(0.01,0.73),
              xycoords='axes fraction', verticalalignment='bottom', fontsize=8)

nextSP() 
field     = heat_restore
field_min = numpy.amin(field)
field_max = numpy.amax(field)
field_ave = (field*area).sum() / area.sum()
ch = plt.pcolormesh(lon,lat,field)
cbax=plt.colorbar(ch, extend='both')
plt.title('heat_restore[$W/m^2$]') 
plt.clim(-20,20)
plt.xlim(-280,80)
plt.ylim(-80,90)
plt.ylabel(r'Latitude [$\degree$N]')
plt.xlabel(r'Longitude')
plt.gca().set_axis_bgcolor('gray')
axis = plt.gca()
axis.annotate('max=%5.2f\nmin=%5.2f\nave=%5.2f'%(field_max,field_min,field_ave),xy=(0.01,0.73), 
              xycoords='axes fraction', verticalalignment='bottom', fontsize=8)

nextSP() 
field = net_heat_surface-net_heat_coupler-frazil-heat_restore-heat_pme 
residual = (field*area).sum() / (net_heat_surface[n]*area).sum() 
field_min = numpy.amin(field)
field_max = numpy.amax(field)
field_ave = (field*area).sum() / area.sum()
plt.pcolormesh(lon,lat,field)
plt.colorbar()
plt.title('Residual(error)[$W/m^2$]') 
plt.xlim(-280,80)
plt.ylim(-80,90)
plt.ylabel(r'Latitude [$\degree$N]')
plt.xlabel(r'Longitude')
plt.gca().set_axis_bgcolor('gray')
axis = plt.gca()
axis.annotate('nondim res=%5.1e'%(residual),xy=(0.01,0.84), 
              xycoords='axes fraction', verticalalignment='bottom', fontsize=8)
```

```python
# Heat fluxes crossing ocean surface via the coupler
# net_heat_coupler =  LwLatSens + SW 
# LwLatSens = LW + Latent + Sensible 

LwLatSens = forcing.variables['LwLatSens'][n]
SW = forcing.variables['SW'][n]

figure(figsize=(16,10))
newSP(2,2);

field     = net_heat_coupler
field_min = numpy.amin(field)
field_max = numpy.amax(field)
field_ave = (field*area).sum() / area.sum()
ch = plt.pcolormesh(lon,lat,field)
cbax=plt.colorbar(ch, extend='both')
plt.title('net_heat_coupler[$W/m^2$]') 
plt.clim(-200,200)
plt.xlim(-280,80)
plt.ylim(-80,90)
plt.ylabel(r'Latitude [$\degree$N]')
plt.xlabel(r'Longitude')
plt.gca().set_axis_bgcolor('gray')
axis = plt.gca()
axis.annotate('max=%5.2f\nmin=%5.2f\nave=%5.2f'%(field_max,field_min,field_ave),xy=(0.01,0.73), 
              xycoords='axes fraction', verticalalignment='bottom', fontsize=8)

nextSP() 
field     = LwLatSens
field_min = numpy.amin(field)
field_max = numpy.amax(field)
field_ave = (field*area).sum() / area.sum()
ch = plt.pcolormesh(lon,lat,field)
cbax=plt.colorbar(ch, extend='both')
plt.title('LwLatSens[$W/m^2$]') 
plt.clim(-200,200)
plt.xlim(-280,80)
plt.ylim(-80,90)
plt.ylabel(r'Latitude [$\degree$N]')
plt.xlabel(r'Longitude')
plt.gca().set_axis_bgcolor('gray')
axis = plt.gca()
axis.annotate('max=%5.2f\nmin=%5.2f\nave=%5.2f'%(field_max,field_min,field_ave),xy=(0.01,0.73), 
              xycoords='axes fraction', verticalalignment='bottom', fontsize=8)

nextSP() 
field     = SW
field_min = numpy.amin(field)
field_max = numpy.amax(field)
field_ave = (field*area).sum() / area.sum()
ch = plt.pcolormesh(lon,lat,field)
cbax=plt.colorbar(ch, extend='both')
plt.title('SW[$W/m^2$]') 
plt.clim(-200,200)
plt.xlim(-280,80)
plt.ylim(-80,90)
plt.ylabel(r'Latitude [$\degree$N]')
plt.xlabel(r'Longitude')
plt.gca().set_axis_bgcolor('gray')
axis = plt.gca()
axis.annotate('max=%5.2f\nmin=%5.2f\nave=%5.2f'%(field_max,field_min,field_ave),xy=(0.01,0.73), 
              xycoords='axes fraction', verticalalignment='bottom', fontsize=8)


nextSP() 
field = net_heat_coupler - SW - LwLatSens
residual = (field*area).sum() / (net_heat_surface[n]*area).sum() 
field_min = numpy.amin(field)
field_max = numpy.amax(field)
field_ave = (field*area).sum() / area.sum()
plt.pcolormesh(lon,lat,field)
plt.colorbar()
plt.title('Residual(error)[$W/m^2$]') 
plt.xlim(-280,80)
plt.ylim(-80,90)
plt.ylabel(r'Latitude [$\degree$N]')
plt.xlabel(r'Longitude')
plt.gca().set_axis_bgcolor('gray')
axis = plt.gca()
axis.annotate('nondim res=%5.1e'%(residual),xy=(0.01,0.84), 
              xycoords='axes fraction', verticalalignment='bottom', fontsize=8, color='white')
```

<h2 align="center">Total heat budget</h2>

We now compute the difference between the net heat crossing the ocean boundaries over a given period to the seawater's heat change over the same time period.  Difference between the two should be at the level of truncation error.  If larger than truncation error, then there is a bug.

The net heat per time (units of Watts) entering through the ocean boundaries is given by the area integral
$$\begin{equation*}
\mbox{boundary heating of liquid seawater} = \int Q \, dA,
\end{equation*}$$
where the net heat flux (units of $\mbox{W}~\mbox{m}^{-2}~\mbox{s}^{-1}$) is given by 
$$\begin{align*}
 Q &= {\tt (net\_heat\_coupler + heat\_pme + frazil) + internal\_heat + heat\_restore} \\
   &= {\tt net\_heat\_surface + internal\_heat + heat\_restore}
\end{align*}$$
The time change of liquid seawater heat is computed according to 
$$\begin{equation*}
\mbox{seawater heat content change} = 
\frac{C_p }{\tau_{n+1} - \tau_{n} } \int dA \left(\rho_0 \int (\Theta_{n+1} - \Theta_{n}) \, \mathrm{d}z \right) 
\end{equation*}$$
 where $\tau_{n+1} - \tau_{n}$ is the time increment in seconds.  Note that we make use of the MOM6 diagnostic for depth integrated potential/conservative temperature  
$$\begin{equation*}
 {\tt tomint} = \rho_0 \int \Theta \, dz.
\end{equation*}$$


```python
tomint = surface.variables['tomint']
time   = surface.variables[tvar][:]*86400.0

n0      = n-1
dtomint = tomint[n] - tomint[n0]
dt      = time[n] - time[n0]
lhs     = cp * area * dtomint / dt
rhs     = area * ( net_heat_coupler + heat_pme + geothermal + frazil + heat_restore )

#print dt*(area*net_heat_coupler[n]).sum(), dt*(area*heat_pme[n]).sum(),dt*(area*geothermal[n]).sum(), dt*(area*frazil[n]).sum() #, dt*(area*heat_restore[n]).sum()
print 'Total seawater heat content change =',dt*lhs.sum(), '[Joules]'
print 'Net heat through boundaries        =',dt*rhs.sum(),'[Joules]'
print 'Residual                           =',dt*lhs.sum() - dt*rhs.sum(),'[Joules]'
print 'Nondimensional residual            =',( lhs.sum() - rhs.sum() )/lhs.sum()
```

### Total salt budget

$$
\frac{10^{-3} \rho_0 }{t(n+1) - t(n)} \int dA \int (S_{n+1} - S_{n}) dz = \int F_s dA
$$
where $F_s =$ `salt_flux` + `salt_flux_restore`

and the variable `somint` = $\rho_0 \int S dz$.

```python
somint = surface.variables['somint']
n0 = n-1
dsomint = somint[n] - somint[n0]
time = surface.variables[tvar][:]*86400.
dt = time[n] - time[n0]
lhs = 1.e-3 * area * dsomint / dt
rhs = area * ( salt_flux ) #+ salt_flux_restore )
#print dt*(area*net_heat_coupler[n]).sum(), dt*(area*heat_pme[n]).sum(),dt*(area*geothermal[n]).sum(), dt*(area*frazil[n]).sum() #, dt*(area*heat_restore[n]).sum()
print 'Total salt change =',dt*lhs.sum(), 'Net salt through boundaries =',dt*rhs.sum(),'[kg of salt]'
print 'Nondimensional residual=',( lhs.sum() - rhs.sum() )/lhs.sum()
```
