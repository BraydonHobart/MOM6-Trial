```python
%pylab inline
import matplotlib.pyplot as plt
import netCDF4
pylab.rcParams['figure.figsize'] = (18.0, 10.0)
```

```python
path = '/work/aja/GOLD_SIS.orig/'
wet = netCDF4.Dataset(path+'ocean_static.nc').variables['wet'][:]
area = netCDF4.Dataset(path+'ocean_static.nc').variables['area_t'][:]*wet
forcing = netCDF4.Dataset(path+'forcing.nc')
surface = netCDF4.Dataset(path+'ocean_sfc.nc') ; tvar = 'time'
cp = 3925.0
rho0 = 1035.
for v in forcing.variables: print v,
for v in surface.variables: print v,
n = 4
```

```python
def newSP(y,x):
    global __spv, __spi ; __spv = (y,x) ; __spi = 1 ; plt.subplot(__spv[0], __spv[1], __spi)
def nextSP():
    global __spv, __spi ; __spi = __spi + 1 ; plt.subplot(__spv[0], __spv[1], __spi)
```

# Surface heat fluxes
net_heat_surface = heat_pme + frazil + net_heat_coupler

```python
net_heat_surface = forcing.variables['net_heat_surface'][n]
net_heat_coupler = forcing.variables['net_heat_coupler'][n]
heat_pme = forcing.variables['Heat_PmE'][n]
frazil = forcing.variables['frazil'][n]
newSP(2,2); plt.pcolormesh( net_heat_surface ) ; plt.colorbar(); plt.title('net_heat_surface = $Q_{net}$ [$W/m^2$]')
nextSP(); plt.pcolormesh( net_heat_coupler ) ; plt.colorbar(); plt.title('net_heat_coupler =$Q_{SW+LW+LH+SH}$ [$W/m^2$]')
nextSP(); plt.pcolormesh( frazil ) ; plt.colorbar(); plt.title('frazil = $Q_{frazil}$ [$W/m^2$]')
nextSP(); plt.pcolormesh( heat_pme ) ; plt.clim((-50,50)); plt.colorbar(); plt.title('heat_pme = $Q_{PRC-E}$ [$W/m^2$]')
```

### Consistent aggregated fluxes

```python
newSP(2,2); plt.pcolormesh( net_heat_coupler + heat_pme + frazil
               - net_heat_surface ) ; plt.colorbar(); plt.title('Residual = coupler + pme + frazil - surface [$W/m^2$]')
print 'Non-dimensional residual ',(( net_heat_coupler + heat_pme + frazil
               - net_heat_surface )*area).sum() / ( net_heat_surface[n]*area ).sum()
```

### Total heat budget

$$
\frac{c_p \rho_0 }{t(n+1) - t(n)} \int dA \int (T_{n+1} - T_{n}) dz = \int Q dA
$$
where $Q =$ ( `net_heat_coupler` + `heat_pme` + `frazil` )  + `internal_heat`
or $Q =$ `net_heat_surface`  + `internal_heat`

and the variable `tomint` = $\rho_0 \int T dz$.

```python
tomint = surface.variables['tomint']
n0 = n -1
dtomint = tomint[n] - tomint[n0]
time = surface.variables[tvar][:]*86400.
dt = time[n] - time[n0]
lhs = cp * area * dtomint / dt
rhs = area * ( net_heat_coupler + heat_pme + geothermal + frazil )# + heat_restore )
#print dt*(area*net_heat_coupler[n]).sum(), dt*(area*heat_pme[n]).sum(),dt*(area*geothermal[n]).sum(), dt*(area*frazil[n]).sum() #, dt*(area*heat_restore[n]).sum()
print 'Total heat change =',dt*lhs.sum(), 'Net heat through boundaries =',dt*rhs.sum(),'[Joules]'
print 'Nondimensional residual=',( lhs.sum() - rhs.sum() )/lhs.sum()
```

# Salt fluxes

```python
salt_flux = forcing.variables['salt_flux'][n]
salt_flux_restore = forcing.variables['salt_flux_restore'][n]
newSP(2,2); plt.pcolormesh( 86400.*salt_flux ) ; plt.colorbar(); plt.title('$F_{s}$ [$kg/m^2/day$]')
nextSP() ; plt.pcolormesh( 86400.*salt_flux_restore ) ; plt.colorbar(); plt.title('$F_{restore}$ [$kg/m^2/day$]')
```

### Total salt budget

$$
\frac{10^{-3} \rho_0 }{t(n+1) - t(n)} \int dA \int (S_{n+1} - S_{n}) dz = \int F_s dA
$$
where $F_s =$ `salt_flux` + `salt_flux_restore`

and the variable `somint` = $\rho_0 \int S dz$.

```python
somint = surface.variables['somint']
n0 = n-1
dsomint = somint[n] - somint[n0]
time = surface.variables[tvar][:]*86400.
dt = time[n] - time[n0]
lhs = 1.e-3 * area * dsomint / dt
rhs = area * ( salt_flux + salt_flux_restore )
#print dt*(area*net_heat_coupler[n]).sum(), dt*(area*heat_pme[n]).sum(),dt*(area*geothermal[n]).sum(), dt*(area*frazil[n]).sum() #, dt*(area*heat_restore[n]).sum()
print 'Total salt change =',dt*lhs.sum(), 'Net salt through boundaries =',dt*rhs.sum(),'[kg of salt]'
print 'Nondimensional residual=',( lhs.sum() - rhs.sum() )/lhs.sum()
```

### Total mass budget

$$
\frac{1}{t(n+1) - t(n)} \int dA \int \left(\rho_{n+1} - \rho_{n} \right) dz = \int \left( P+R+C-E \right) dA
$$
where $P+R+C-E =$ `salt_flux` + `salt_flux_restore`

and the variable `mass_wt` = $\int \rho  dz$.

```python
mass_wt = surface.variables['mass_wt']
n0 = n-1
dmass_wt = mass_wt[n] - mass_wt[n0]
time = surface.variables[tvar][:]*86400.
dt = time[n] - time[n0]
lhs = area * dmass_wt / dt
rhs = area * ( PRCmE )
#print dt*(area*net_heat_coupler[n]).sum(), dt*(area*heat_pme[n]).sum(),dt*(area*geothermal[n]).sum(), dt*(area*frazil[n]).sum() #, dt*(area*heat_restore[n]).sum()
print 'Total mass change =',dt*lhs.sum(), 'Net mass through boundaries =',dt*rhs.sum(),'[kg]'
print 'Nondimensional residual=',( lhs.sum() - rhs.sum() )/lhs.sum()
```
